диаграмма в виде графика для проверки целостности диапазона данных

```python
import pandas as pd
import re
from rapidfuzz import process 

# Список ключевых слов YouTube
youtube_keywords = [
    'ютуб', 'youtube', 'ютб', 'youtub', 'ютюб', 'ютюбе', 'ютубе',
    'you tube', 'youtubе', 'ютуб.', 'youtube.com', 'ютуб.ком'
]

def is_youtube_query(query):
    """Проверяет, содержит ли запрос ключевые слова YouTube"""
    query_str = str(query).lower()
    return any(kw in query_str for kw in youtube_keywords)

def find_best_match(query, choices, threshold=60):
    if not is_youtube_query(query):
        return None
    # rapidfuzz.process.extractOne возвращает (match, score, index)
    result = process.extractOne(query, choices)
    if result is None:
        return None
    best_match, score, _ = result
    return best_match if score >= threshold else None

# Стандартные значения
youtube_standard = pd.DataFrame({'standard': ['ютуб', 'youtube']})
choices = youtube_standard['standard'].tolist()

# Применяем
df['youtube_match'] = df['query'].apply(lambda x: find_best_match(x, choices, threshold=60))
df['Query_Fuzzy'] = df['youtube_match'].fillna(df['query'])

# Экспорт
df[['query','device', 'youtube_match', 'Query_Fuzzy']].to_csv('fuzzy_youtube_clean.csv', index=False, encoding='utf-8-sig')
print("✅ Экспорт завершён: fuzzy_youtube_clean.csv")
```

```python
from pandasql import sqldf

query_pivot = """
SELECT
    Query_Fuzzy,
    query,
    SUM(CASE WHEN device = 'desktop' THEN 1 ELSE 0 END) AS desktop,
    SUM(CASE WHEN device = 'touch' THEN 1 ELSE 0 END) AS touch
FROM df
WHERE youtube_match IS NOT NULL
GROUP BY Query_Fuzzy, query
ORDER BY desktop DESC, touch DESC;
"""

result_pivot = sqldf(query_pivot)
result_pivot
```
